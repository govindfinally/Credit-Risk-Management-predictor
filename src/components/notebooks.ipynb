```python
def plot_correlation_heatmap(df):
    plt.figure(figsize=(20,15))
    sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
    plt.title('Correlation Heatmap')
    plt.show()
plot_correlation_heatmap(df.head(10))
# checking the correlation between the numerical columns
```

```python
def removing_nearly_same_columns(df,perce=0.75):
    columns_corr=set()# to omit duplicated columns also
    corr_matrix=df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>perce:
                colname=corr_matrix.columns[i]
                columns_corr.add(colname)
    return columns_corr
```

```python
print(removing_nearly_same_columns(df,perce=0.8))
```
